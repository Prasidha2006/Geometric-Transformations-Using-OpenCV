# **Image Translation**

Write a Python program using OpenCV to perform image translation on the **"Given image"**.

The program should:

1. Import `cv2`, `numpy`, and `matplotlib.pyplot`.
2. Read the image **"given** in color mode.
3. Create a **translation matrix** that moves the image  
   **50 pixels to the right** and **80 pixels down**.
4. Apply the translation
5. Display the following using a **1 × 2 layout** in Matplotlib:  
   - **Original Image**  
   - **Translated Image**


```python
# Import Required Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the image using OpenCV
img = cv2.imread('baseball.jpg')

# Convert BGR to RGB for displaying in Matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Get image dimensions
rows, cols = img.shape[:2]
```

```python
# Create Translation Matrix
tx = 50   # shift right
ty = 80   # shift down

translation_matrix = np.float32([
    [1, 0, tx],
    [0, 1, ty]
])
```

```python
# Apply Translation
translated_img = cv2.warpAffine(img_rgb, translation_matrix, (cols, rows))
```

```python
# Display Images
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(translated_img)
plt.title("Translated Image")
plt.axis("off")

plt.show()
```

# **Image Shearing**

Write a Python program using OpenCV to perform image shearing on the **given image**.

The program should:

1. Create a **shearing transformation matrix** to apply:
   - **Horizontal shearing**  
   - **Vertical shearing**
2. Apply the shearing transformations.
3. Display the following using a **1 × 3 layout** in Matplotlib:  
   - **Original Image**  
   - **Horizontally Sheared Image**  
   - **Vertically Sheared Image**


```python
# Horizontal Shearing Matrix
horizontal_shear = np.float32([
    [1, 0.5, 0],
    [0, 1, 0]
])
```

```python
# Vertical Shearing Matrix
vertical_shear = np.float32([
    [1, 0, 0],
    [0.5, 1, 0]
])
```

```python
# Apply Shearing
horizontal_sheared = cv2.warpAffine(img_rgb, horizontal_shear, (int(cols*1.5), rows))

vertical_sheared = cv2.warpAffine(img_rgb, vertical_shear, (cols, int(rows*1.5)))
```

```python
# Display Images
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(horizontal_sheared)
plt.title("Horizontally Sheared")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(vertical_sheared)
plt.title("Vertically Sheared")
plt.axis("off")

plt.show()
```

# **Image Reflection**

Write a Python program using OpenCV to perform image reflection on the **given image**.

The program should:

1. Generate the following reflected versions of the image:
   - **Horizontal Reflection** (flip left ↔ right)  
   - **Vertical Reflection** (flip top ↔ bottom)

# **Image Rotation**

Write a Python program using OpenCV to perform **image rotation** on the **given image**.

The program should:

1. Rotate the image by:  
   - **45 degrees**  
   - **90 degrees**
2. Use `cv2.getRotationMatrix2D` and `cv2.warpAffine`.
3. Display the following using a **1 × 3 layout** in Matplotlib:  
   - **Original Image**  
   - **45° Rotated Image**  
   - **90° Rotated Image**


```python
# Rotation by 45 Degrees
rotation_matrix_45 = cv2.getRotationMatrix2D((cols/2, rows/2), 45, 1)

rotated_45 = cv2.warpAffine(img_rgb, rotation_matrix_45, (cols, rows))
```

```python
# Rotation by 90 Degrees
rotation_matrix_90 = cv2.getRotationMatrix2D((cols/2, rows/2), 90, 1)

rotated_90 = cv2.warpAffine(img_rgb, rotation_matrix_90, (cols, rows))
```

```python
# Display Images
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(rotated_45)
plt.title("45° Rotated Image")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(rotated_90)
plt.title("90° Rotated Image")
plt.axis("off")

plt.show()
```
